```python
import pandas as pd
import matplotlib.pyplot as plt

link="https://github.com/seismomat/HerramientasComputacionales/blob/main/Aplicaciones/ebola_outbreak_dataset-master/ebola_outbreak_dataset-master/ebola_2014_2016_clean.csv"
code="?raw=true"
url=link+code
df=pd.read_csv(url,sep=",",index_col=1)
```

```python
SL=df[df["Country"]=="Sierra Leone"]
```

```python
cuantos=df.groupby("Country").size().reset_index(name="Conteo")
cuantos
```

```python
plt.pie(cuantos["Conteo"],labels=cuantos["Country"],autopct="%0.1f%%")
plt.axis("equal")
plt.title("Distribución  del ébola por países")
plt.savefig("Patel_Paises.png",dpi=300,format="png")
plt.show()
```

```python
mean_values=SL["No. of confirmed deaths"].mean()
mean_values
```

```python
SL.fillna(mean_values,inplace= True)
```

```python
SL.isnull().sum()
```

```python
Dates=SL.index[0:-1:10]
NCC=SL["No. of confirmed cases"][0:-1:10]
NCD=SL["No. of confirmed deaths"][0:-1:10]
```

```python
plt.plot(Dates,NCC,"*--m",ms=2,label="infectados")
plt.plot(Dates,NCD,"d--k",ms=2,label="Decesos")
plt.grid()
plt.title("Avance del ébola en Sierra Leona")
plt.xlabel("periodo de tiempo")
plt.ylabel("Casos confirmados")
plt.xticks(rotation=90,fontsize=8)
plt.legend()
plt.show()
```

```python
fig=plt.figure(figsize=(12,10))
plt.plot(Dates,NCC,"*--y",ms=2,label="infectados")
plt.bar(Dates,NCC,color="yellow",alpha=0.2)
plt.plot(Dates,NCD,"d--g",ms=2,label="Decesos")
plt.bar(Dates,NCD,color="green",alpha=0.2)
plt.grid()
plt.title("Avance del ébola en Sierra Leona")
plt.xlabel("periodo de tiempo")
plt.ylabel("Casos confirmados")
plt.xticks(rotation=90,fontsize=8)
plt.savefig("Avance_de_lébola_en_SL.png",dpi=300,format="png")
plt.legend()
plt.show()
```

```python
PIvsD=100*SL["No. of confirmed deaths"]/SL['No. of confirmed cases']
SL["PIvsD"]=PIvsD
```

```python
SL.head()
```

```python
recuperados=((100-PIvsD)/100)*SL['No. of confirmed cases']
SL["recuperados"]=recuperados
SL.head()
```

```python
SL["PR"]=(100-PIvsD)
SL.head()
```

```python
PIsvD_mean=SL["PIvsD"].mean()
PR_mean=SL["PR"].mean()
porcentajes=[PIsvD_mean,PR_mean]
etiquetas=["PIvsD","PR"]
```

```python
plt.pie(porcentajes,labels=etiquetas,autopct="%0.1f%%",colors=["green","yellow"])
plt.axis("equal")
plt.title("Porcentaje de Recuperados contra Decesos")
plt.show()
```

```python
G=SL["Girls"][0:-1:15]
AF=SL["Adult Female"][0:-1:15]
OF=SL["Old Female"][0:-1:15]
Dates=SL.index[0:-1:15]
fig=plt.figure(figsize=(12,6))
plt.bar(Dates,AF,color='red',alpha=0.8,label='Niñas')
plt.bar(Dates,G,color='blue',alpha=0.8,label='Adultas')
plt.bar(Dates,OF,color='yellow',alpha=0.8,label='Adultas Mayores')
plt.grid()
plt.title("Distribución del ébola en mujeres",fontsize=15)
plt.ylabel("Casos confirmados")
plt.xticks(rotation=90,fontsize=8)
plt.legend(loc='upper left',bbox_to_anchor=(1.02,0.5))
plt.savefig("FemaleSL.png",dpi=300,format="png")
plt.show()
```

```python
Boys=SL['Boys'].to_numpy()
SL_Boys=pd.DataFrame({'Boys':Boys})
SL_Boys.index=pd.to_datetime(SL.index,format='%d/%m/%Y')
SL_Boys=SL_Boys.sort_index()
```

```python
SL_Boys
```

```python
steps=57

data_train=SL_Boys[:-steps]
data_test=SL_Boys[-steps:]
```

```python
print(f"Las fechas  son {len((data_train))}")
```

```python
fig,ax=plt.subplots(figsize=(8,3))
data_train['Boys'].plot(ax=ax,label='train')
data_test['Boys'].plot(ax=ax,label='test')
plt.show()
```

```python
from skforecast.ForecasterAutoreg import ForecasterAutoreg
from sklearn.ensemble import RandomForestRegressor
```

```python
forecaster=ForecasterAutoreg(
    regressor=RandomForestRegressor(random_state=123),
    lags=6
)
```

```python
forecaster.fit(y=data_train['Boys'])
```

```python
steps=57
predicciones=forecaster.predict(steps)
```

```python
predicciones.index=data_test.index
```
